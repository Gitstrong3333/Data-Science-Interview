# **Which function is commonly used to convert a multidimensional NumPy array into a one-dimensional array? If the resulting array is modified, will it affect the original array? Explain your answer.**

The flatten() function in NumPy is commonly used to convert a multidimensional array into a one-dimensional array.

The flatten() method returns a copy of the original array. Therefore, if you modify the flattened array, it will not affect the original array.

```python
import numpy as np

# Create a 2D array
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Original Array:")
print(arr)

# Convert to 1D using flatten()
flat_arr = arr.flatten()

print("\nFlattened Array:")
print(flat_arr)

# Modify flattened array
flat_arr[0] = 100

print("\nModified Flattened Array:")
print(flat_arr)

print("\nOriginal Array After Modification:")
print(arr)
```

# **2.If two variables are defined as a = 3 and b = 4, will the id() function return the same value for both variables? Explain your answer.**



Answer: Rewritten Answer

No, the id() function will not return the same value for a and b.

In Python, the id() function returns the unique identity of an object, which typically corresponds to its memory address. Since 3 and 4 are two different integer objects, they have different identities in memory. Therefore, id(a) and id(b) will return different values.

Each object in Python has a unique identity that remains constant during its lifetime.

```python
a = 3
b = 4

print(id(a))
print(id(b))
```

# **3.If two variables are defined as mean = 8 and Mean = 8, are they considered the same variable in Python? Explain your answer.**


No, mean and Mean are not considered the same variable in Python.

Python is a case-sensitive programming language, which means it treats uppercase and lowercase letters as different characters. Therefore, variable names that differ only in case are treated as completely separate variables.

Even though both variables store the same value (8), they are distinct identifiers in memory.

```python
mean = 8
Mean = 8

print(mean)
print(Mean)
```

Both will print 7, but they are two different variables.

# **4.What is the purpose of the inplace parameter in Pandas functions? Explain its behavior.**


The inplace parameter in Pandas functions determines whether a DataFrame is modified directly or a modified copy is returned.

When inplace=True, the operation is performed directly on the original DataFrame, and the function returns None.

When inplace=False (default), the function returns a new modified DataFrame, and the original DataFrame remains unchanged.

```python
import pandas as pd

df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

# Using inplace = False (default)
df2 = df.drop('A', axis=1)
print("Original DataFrame:")
print(df)

print("\nNew DataFrame:")
print(df2)

#Now Using inplace=True

df.drop('A', axis=1, inplace=True)
print(df)
```

Important Interview Note

In modern Pandas usage, many developers prefer not using inplace=True because:

It does not always provide performance benefits.

It can sometimes make debugging harder.

Chaining operations becomes easier when returning a new DataFrame.

# **5.How can you change the index of a DataFrame in Python (Pandas)? Explain the method used.**

In Pandas, you can change the index of a DataFrame using the set_index() method.

🔹 Syntax:
DataFrame.set_index(keys, drop=True, append=False, inplace=False, verify_integrity=False)
🔹 Explanation of Important Parameters:

keys: Column name(s) or array-like object(s) to set as the new index.
This can be:

A single column name

A list of column names (for multi-index)

An array, Series, or NumPy array of the same length as the DataFrame

drop (default = True):
Removes the column(s) used as the index from the DataFrame.

inplace (default = False):
If True, modifies the original DataFrame. Otherwise, returns a new DataFrame.

append (default = False):
If True, adds the new index to the existing index instead of replacing it.

verify_integrity (default = False):
Checks for duplicate index values.

```python
# Example 1: Setting a Single Column as Index

import pandas as pd

df = pd.DataFrame({
    'ID': [101, 102, 103],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Score': [85, 90, 95]
})

df_new = df.set_index('ID')
print(df_new)

#Example 2: Changing Index Using inplace=True

df.set_index('Name', inplace=True)
print(df)

#Example 3: Setting Multiple Columns (MultiIndex)

df = pd.DataFrame({
    'Department': ['IT', 'IT', 'HR'],
    'Employee': ['A', 'B', 'C'],
    'Salary': [50000, 60000, 45000]
})

df_multi = df.set_index(['Department', 'Employee'])
print(df_multi)
```

# **6.How can you check whether a number is prime or not in Python?**


A prime number is a number greater than 1 that is divisible only by 1 and itself.

To check whether a number is prime in Python, we test if it is divisible by any number other than 1 and itself. If it is divisible by any other number, it is not prime.

```python
num = int(input("Enter any number: "))

if num > 1:
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
else:
    print(num, "is not a prime number")

    #Why use int(num**0.5) + 1?

#We only need to check divisibility up to the square root of the number. This makes the solution more efficient.
```

# **8.How does a for loop differ from a while loop in Python? Explain with examples.**


Both for and while loops are used to execute a block of code repeatedly, but they differ in how the repetition is controlled.

🔹 for Loop

# **9.How can you import multiple Excel sheets into a single DataFrame in Python? Explain with an example.**

In Python, you can use the pd.read_excel() function from the Pandas library to read multiple sheets from an Excel file.

If you want to load all sheets at once, you can set the parameter sheet_name=None. This returns a dictionary where:

Keys → Sheet names

Values → Corresponding DataFrames

You can then combine all sheets into a single DataFrame using pd.concat().



```python
import pandas as pd

# Read all sheets from the Excel file
excel_file = pd.read_excel("data.xlsx", sheet_name=None)

# Combine all sheets into one DataFrame
df = pd.concat(excel_file.values(), ignore_index=True)

print(df)

```

#Explanation

sheet_name=None → Reads all sheets.

excel_file.values() → Extracts all DataFrames from the dictionary.

pd.concat(..., ignore_index=True) → Combines them into one DataFrame and resets the index.




```python
#If You Want Specific Sheets Only:

df = pd.concat(
    pd.read_excel("data.xlsx", sheet_name=["Sheet1", "Sheet2"]),
    ignore_index=True
)
```

# **Interview Tip for Question 9:**

sheet_name="Sheet1" → Load one sheet.

sheet_name=["Sheet1","Sheet2"] → Load selected sheets.

# **10.What is the difference between the append() and extend() methods in Python lists? Explain with examples.**



Both append() and extend() are list methods in Python used to add elements to a list. However, they work differently.

🔹 append() Method

The append() method adds a single element to the end of the list.
If you append a list, the entire list is added as one element.

Syntax:
list.append(item)


```python
list1 = [1, 2, 3]
list1.append([4, 5])

print(list1)
```

Here, [4, 5] is added as a single element.

**extend() Method**

The extend() method adds each element of an iterable (like a list, tuple, or string) to the list individually.

Syntax:
list.extend(iterable)
Example:

```python
list2 = [1, 2, 3]
list2.extend([4, 5])

print(list2)
```

| `append()`                      | `extend()`                                        |
| ------------------------------- | ------------------------------------------------- |
| Adds one element                | Adds multiple elements                            |
| Can add a list as a single item | Adds elements of the list individually            |
| Increases list length by 1      | Increases list length by number of elements added |
